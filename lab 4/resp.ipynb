```python
import requests
import json
```

```python
raw = {
    "message": "посоветуй как мне стоит питаться, чтобы сбросить вес"
}

url = "http://localhost:8000/generate"

query = requests.post(url, json=raw)

data = json.loads(query.text)

success = data["success"]
msg = data["message"]
exer = data["exercises"]
nut = data["nutrition"]
print(f"Статус - {success}\n\nСообщения - {msg}\n\nУпражнения - {exer}\n\nNutrition - {nut}")
```
