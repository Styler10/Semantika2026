```python
import requests
import json
```

```python
raw = {
    "message": "посоветуй как мне стоит питаться, чтобы сбросить вес"
}

url = "http://localhost:8000/generate"

query = requests.post(url, json=raw)

data = json.loads(query.text)
text = data["response"]
print(text)
```
